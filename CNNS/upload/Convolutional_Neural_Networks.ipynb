# Convolutional Neural Networks

## Convolutional Filters

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_classes = 10
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_width = x_train.shape[1]
x_height = x_train.shape[2]
np.random.seed(2)

x_val = x_train[:999,...]
y_val = y_train[:999,...]

x_train = x_train[1000:,...]
y_train = y_train[1000:,...]

@tf.function
def normalize_img(data, label):
    return tf.cast(data, tf.float32) / 255., label
    
@tf.function
def one_hot_encoding(data, label):
    one_hot_labels = tf.one_hot(label, depth=n_classes)
    return data, one_hot_labels


dataset_train = tf.data.Dataset.from_tensor_slices((x_train, y_train))
dataset_train = dataset_train.map(normalize_img)
dataset_train = dataset_train.map(one_hot_encoding)
dataset_train = dataset_train.cache()
dataset_train = dataset_train.shuffle(x_train.shape[0])
dataset_train = dataset_train.batch(64)
dataset_train = dataset_train.prefetch(tf.data.AUTOTUNE)

dataset_val = tf.data.Dataset.from_tensor_slices((x_val, y_val))
dataset_val = dataset_val.map(normalize_img)
dataset_val = dataset_val.map(one_hot_encoding)
dataset_val = dataset_val.batch(64)
dataset_val = dataset_val.cache()
dataset_val = dataset_val.prefetch(tf.data.AUTOTUNE)

dataset_test = tf.data.Dataset.from_tensor_slices((x_test, y_test))
dataset_test = dataset_test.map(normalize_img)
dataset_test = dataset_test.map(one_hot_encoding)
dataset_test = dataset_test.batch(64)
dataset_test = dataset_test.cache()
dataset_test = dataset_test.prefetch(tf.data.AUTOTUNE)
```

## Convolutional Filters

Convolutional Filters are a central component for Convolutional Neural Networks (hence the name). They are shifted in two dimensions, x and y, of an input image. For each (x, y) position all pixels in the filter range are multiplied with their respective filter coefficient and summed up to yield a single output. This way, pixel by pixel, the convolutional filter generates a feature map.

```python
def conv_kernel(image, kernel):
    # determining the kernel range
    k_half_width = int(kernel.shape[0] / 2)
    k_half_height = int(kernel.shape[1] / 2)
    result_image = np.zeros_like(image, dtype=float)
    # iterating over all image pixels in two dimensions
    for y in range(k_half_height, image.shape[1] - k_half_height):
        for x in range(k_half_width, image.shape[0] - k_half_width):
            # iterating over all pixel coefficients in two dimensions
            for y_k in range(kernel.shape[1]):
                for x_k in range(kernel.shape[0]):
                    # Multiply pixel value with coefficient and add it to the result pixel
                    # TODO
                    result_image[x, y] += 1
    return result_image
```

Comvolutional filters can be used to extract features from images, such as color, edges, points and so on. There are several popular filters known in the area of image processing, as the following examples show. Later however, the goal to train these filters using back propagation instead of setting them analytically.

```python
# This filter represents a moving average and smoothes the image.
# If the sum of all coeficients of the filter yields a number greater zero, it can be interpreted as low pass filter
kernel_average = 1. / 9. * np.array([[1, 1, 1],
                                     [1, 1, 1],
                                     [1, 1, 1]])

# This one detects vertical edges in the image. The sum of its coefficients is zero, hence, it can be interpreted as high pass
# The filter also represents the first derivative in the veritical direction
kernel_right_edge = np.array([[-1, 1, 0],
                              [-1, 1, 0],
                              [-1, 1, 0]])

# Detects edges as above but in the diagonal direction
kernel_diag_edge = np.array([[-1,-1, 1],
                             [-1, 1, 0],
                             [ 1, 0, 0]])

# detects points, and can be interpreted as second derivative along two directions
kernel_point = np.array([[-1,-1,-1],
                         [-1, 8,-1],
                         [-1,-1,-1]])

# Another edge fitler, the so called Sobel Filter
kernel_sobel = np.array([[-1,-2,-1],
                         [ 0, 0, 0],
                         [ 1, 2, 1]])
image = x_train[2]

average_map = conv_kernel(image, kernel_right_edge)
right_edge_map = conv_kernel(image, kernel_right_edge)
diag_edge_map = conv_kernel(image, kernel_diag_edge)

point_map = conv_kernel(image, kernel_point)
sobel_map = conv_kernel(image, kernel_sobel)

fig, axes = plt.subplots(3, 3)
fig.set_figheight(10)
fig.set_figwidth(10)

axes[0, 0].matshow(image, cmap="coolwarm")
axes[0, 0].set_title("Original Image")
axes[0, 1].matshow(kernel_average, cmap="coolwarm")
axes[0, 1].set_title("Filter Kernel")
for (i, j), z in np.ndenumerate(kernel_average):
    axes[0, 1].text(j, i, '{:0.2f}'.format(z), ha='center', va='center')
axes[0, 2].matshow(average_map, cmap="coolwarm")
axes[0, 2].set_title("Feature Map")

axes[1, 0].matshow(image, cmap="coolwarm")
axes[1, 1].matshow(kernel_diag_edge, cmap="coolwarm")
for (i, j), z in np.ndenumerate(kernel_diag_edge):
    axes[1, 1].text(j, i, '{:0.1f}'.format(z), ha='center', va='center')
axes[1, 2].matshow(diag_edge_map, cmap="coolwarm")
    
axes[2, 0].matshow(image, cmap="coolwarm")
axes[2, 1].matshow(kernel_point, cmap="coolwarm")
for (i, j), z in np.ndenumerate(kernel_point):
    axes[2, 1].text(j, i, '{:0.1f}'.format(z), ha='center', va='center')
axes[2, 2].matshow(point_map, cmap="coolwarm")
plt.savefig("kernels.png", dpi=200, bbox_inches="tight")
plt.show()
```

## Building a Convolutional Neural Network

A typical CNN contains of stacks of Conv2D and pooling layers. Usually, with each conv Conv2D layer the number of extracted features is increased while with each pooling layer, the lateral resolution is decreased. This way the unstructured input image is transformed step by step to a structured 1D feature vector.

```python
input_layer = tf.keras.layers.Input(shape=(x_width, x_height, 1), name="input_layer")

conv_layer = tf.keras.layers.Conv2D(filters=16,
                                    kernel_size=3,
                                    activation="relu",
                                    name="conv_1")(input_layer)
pooling = tf.keras.layers.MaxPool2D(pool_size=(2, 2),
                                    name="pool_1")(conv_layer)

conv_layer = tf.keras.layers.Conv2D(filters=32,
                                    kernel_size=3,
                                    activation="relu",
                                    name="conv_2")(pooling)
pooling = tf.keras.layers.MaxPool2D(pool_size=(2, 2),
                                    name="pool_2")(conv_layer)

conv_layer = tf.keras.layers.Conv2D(filters=64,
                                    kernel_size=3,
                                    activation="relu",
                                    name="conv_3")(pooling)
pooling = tf.keras.layers.GlobalMaxPool2D(name="global_pool")(conv_layer)


output_layer = tf.keras.layers.Dense(units=10,
                                     activation="softmax",
                                     name="output_layer")(pooling)
```

```python
model = tf.keras.Model(inputs=input_layer, outputs=[output_layer])
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[tf.keras.metrics.CategoricalAccuracy()])
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                 patience=5)

lr_reduction = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',
                                                    factor=0.5,
                                                    patience=2,
                                                    min_lr=1e-4)
history = model.fit(
    dataset_train,
    validation_data=dataset_val,
    epochs=5,
    callbacks=[early_stopping, lr_reduction]
    )

model.save(r"first_conv_model.keras", save_format="keras")
```

## Feature Embedding

```python
example_batch = example_data.next()[0]
batch_out = multi_out.predict(example_batch)
batch_out[6].shape
```

```python
from sklearn.decomposition import PCA
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

pca = PCA(n_components=2)
example_batch_2D = pca.fit_transform(batch_out[6])
print(pca.explained_variance_ratio_)

def imscatter(x, y, images, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    digits = list()
    for x0, y0, im in zip(x, y, images):
        im = OffsetImage(im, zoom=zoom, cmap="gist_gray")
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        digits.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return digits
    
fig, ax = plt.subplots()
imscatter(example_batch_2D[:, 0],
          example_batch_2D[:, 1],
          example_batch,
          zoom=0.5,
          ax=ax)
ax.set_xlabel("1st Main Component")
ax.set_ylabel("2nd Main Component")
plt.show()
```
